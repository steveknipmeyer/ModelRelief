```python
import cv2
import matplotlib.pylab as plt
import numpy as np
import os
```

```python

image = cv2.imread('LucyBackground.png')

kernel = (21, 21)
image_blurred = cv2.GaussianBlur(image, kernel, 0)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

mask = np.zeros(image.shape, np.uint8)
contours, hierarchy = cv2.findContours(gray.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
edge_width = 5
cv2.drawContours(mask, contours, -1, (255,255,255), edge_width)

plt.imshow(mask)

output = np.where(mask==np.array([255, 255, 255]), image_blurred, image)

plt.imshow(output)
gray.shape
```
